In physics, stuff like chemical bonding works because it's just "easier" for molecules to arrange themselves in certain ways than others. We describe how "easy" arragements are using energy, and entropy is the idea that, over time, the universe takes the path of least resistance and moves towards the "easiest" arrangements. This is actually part of the reason why the first parity rule occurs (it's more energy favorable for A's to be across from T's), so what if the second parity rule could be explained by the same principle?

This is, unfortuately, as far as my knowledge can go in the couple of weeks that I have to work on this project. I'm not well-versed enough in the physics to know enough about energy equations to test this for myself, but, thankfully, other researchers have already done that, and those efforts are what I will spend this document talking about.

Piero Fariselli, Cristian Taccioli, Luca Pagani and Amos Maritan derived a series of energy equations to dictate how frequently each sequence and its reverse compliment occurs within a single strand of DNA. Like before, I'm barely well-versed enough in the physics to understamd what these equations mean, let alone evalutae if they're accurate, but suffice it to say that they created equations to represent the energy different arrangements would require, and then used these equations to make predictions. Notably, these equations did predict that we'd observe the second parity rule.

But, they also predicted a bunch of other stuff, too. For one, if it's energy-favorable to include a sequence and its RC together in a single strand, then it should still be energy-favorable to do that with "gapped" RCs, or a reverse complement with a few missing nucleotides. Another prediction is that this should hold for reverse complements, but not regular complements.

The researchers then tested both of these predictions in actual DNA and found that both were correct!

Here, the dark grey represents the frequency of gapped RCs, and the light grey represents the frequency of regular complements. As you can see, there is a clear correlation in the gapped RCs, but no correlation for the regular complements. This adds a lot of credibility to their theory.
